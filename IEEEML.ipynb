### data is not imbalance and not so balance easily can handled

```python
X = train_df.drop('Class',axis=1)
Y = train_df['Class']
```

 ### Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

```python
X_test_df = test_df.drop('ID',axis = 1)
```

### Standardization

```python
from sklearn.preprocessing import StandardScaler
scale =  StandardScaler()

X_train_scaled = scale.fit_transform(X_train)
X_test_scaled  = scale.transform(X_test)
X_test_df = scale.transform(X_test_df)
```

### Model Training - Tabular data

#### logistic regression

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(class_weight='balanced')
penalty = ['l1', 'l2', 'elasticnet']
c_value = [100,10,1,0.1,0.01]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
param = dict(
    penalty = penalty,
    C = c_value,
    solver = solver
)

param
```

```python
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=logistic,param_grid=param,scoring = 'f1',cv = cv,n_jobs=-1)
grid
```

```python
grid.fit(X_train_scaled,Y_train)
```

```python
print(grid.best_score_)
print(grid.best_params_)
```

#### Performence Metrics For Logistic Regression

```python
from sklearn.metrics import confusion_matrix,classification_report,f1_score
Y_pred = grid.predict(X_test_scaled)

print(confusion_matrix(Y_test,Y_pred))
print(classification_report(Y_test,Y_pred))
print(f1_score(Y_test,Y_pred))

```

### Balanced Random forest classification

```python
from imblearn.ensemble import BalancedRandomForestClassifier
```

```python
rf = BalancedRandomForestClassifier(class_weight='balanced', random_state=42)
rf.fit(X_train_scaled, Y_train)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
y_prediction=rf.predict(X_test_scaled)
```

```python
score=accuracy_score(y_prediction,Y_test)
print(score)
```

```python
score2=classification_report(y_prediction,Y_test)
print(score2)
```

### XG Boost Classification

```python
#lets try xgboost
from xgboost import XGBClassifier
from sklearn.metrics import classification_report


# Calculate class weight ratio
scale_pos_weight = Y_train.value_counts()[0] / Y_train.value_counts()[1]

xgb = XGBClassifier(
    scale_pos_weight=scale_pos_weight,  # handles imbalance
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)
```

```python
xgb.fit(X_train_scaled, Y_train)
```

```python
y_pred = xgb.predict(X_test_scaled)
print(classification_report(Y_test, y_pred))
```

```python
score=accuracy_score(y_pred,Y_test)
print(score)
```

### Final Conclusion - We are using XG Boost for prediction . Why....?

Our exploratory analysis revealed significant feature overlap and non-linear boundaries in the dataset. While Balanced Random Forest provided a robust baseline, we chose XGBoost because its gradient boosting framework is better equipped to iteratively minimize residuals in these high-variance regions, ultimately yielding higher precision and superior generalization on our test set. and it is best for tabular data like this.

```python
Pred = xgb.predict(X_test_df)
```
